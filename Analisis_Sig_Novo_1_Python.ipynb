```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.io as pio
import scipy.stats as s
from scipy import signal
```

```python
# Load the .csv files into numpy arrays
Spectrums = np.genfromtxt('Sig_para_Spec_Novo.csv', delimiter=',')
Wavelengths = np.genfromtxt('Sig_para_WL_Novo.csv', delimiter=',')
```

```python
#plot signal with matplotlib
fig, ax0 = plt.subplots(figsize=(10,5))
for i in range(len(Spectrums)):
    ax0.plot(Wavelengths, Spectrums[i])
ax0.set_title('Spectra')
ax0.set_ylabel('Signal [-]')
ax0.set_xlabel('Wavelength [nm]')
ax0.grid(True)
plt.show()
```

```python
#plot signal with plotly (interactive)
fig = go.Figure()

for i in range(len(Spectrums)):
    fig.add_trace(go.Scatter(x=Wavelengths, y=Spectrums[i], name=f'Spectrum {i}'))

fig.update_layout(
    title='Spectra',
    xaxis_title='Wavelength [nm]',
    yaxis_title='Signal [-]',
    width=1000,
    height=500,
    showlegend=True,
    xaxis=dict(showgrid=True, gridwidth=1, gridcolor='lightgray'),
    yaxis=dict(showgrid=True, gridwidth=1, gridcolor='lightgray')
)

pio.show(fig)
```

```python
# Filter the signal, detrend
Spectrums_detrend = signal.detrend(Spectrums, axis=1)
```
